```python
def categorize_col(df, col, labels):

    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]
    df[col] = pd.cut(df[col], edges, labels = labels, duplicates = 'drop')
    return df
```

```python
categorize_col(df, 'track_popularity', ['not_good', 'not_bad', 'average', 'popular'])
df['track_popularity']
```
